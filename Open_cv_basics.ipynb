## Working of Video..

```python
#Establish connection
cap = cv2.VideoCapture(os.path.join('data','traffic.mp4'))
#loop 
for frame in range(int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    #read a frame 
    ret , frame = cap.read()
    # show frame 
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

#close
cap.release()
cv2.destroyAllWindows()
```
